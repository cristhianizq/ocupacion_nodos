```python
use_cols =['NODO', 'CMTS', 'CABLEMAC','FECHA', 'USO','USO_MAX']
df_a_down = pd.read_csv('DOWN/Detalle_(2)_Full_Data_data (19).csv',
                        sep=';', decimal=',', usecols=use_cols)
df_a_up = pd.read_csv('UP/Detalle_Full_Data_data (13).csv',
                        sep=';', decimal=',', usecols=use_cols) 
df_a_down['FECHA'] = [x[0:10] for x in df_a_down['FECHA']]
df_a_up['FECHA'] = [x[0:10] for x in df_a_up['FECHA']]
df_n = pd.read_csv('Prueba_data_nueva_consulta/OCUPACION_COMPLETO.csv')
df_n.columns = ['NODO', 'CMTS', 'CABLEMAC', 'OCUPACION', 'SENTIDO', 'ANIO', 'MES','DIA']
df_n['DIA'] = [str(x) if len(str(x))==2 else '0'+str(x) for x in df_n['DIA']]
df_n['MES'] = [str(x) if len(str(x))==2 else '0'+str(x) for x in df_n['MES']]
df_n['FECHA'] = df_n['DIA'].astype('str') + '/'+ df_n['MES'].astype('str') + '/'  + df_n['ANIO'].astype('str')
df_n = df_n[['NODO', 'CMTS', 'CABLEMAC', 'OCUPACION', 'SENTIDO','FECHA']]
list_cols = ['NODO', 'CMTS', 'CABLEMAC', 'OCUPACION','FECHA']
df_nueva_down = df_n.loc[(df_n['SENTIDO']=='DOWN'),list_cols]
df_nueva_up = df_n.loc[(df_n['SENTIDO']=='UP'),list_cols]
```

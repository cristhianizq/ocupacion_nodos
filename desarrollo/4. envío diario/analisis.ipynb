# Leyendo archivos de bajas 2021

Los nombres de los archivos en "DATA/archivos_bajas_2021/" deben tener la estructura 

Base Cns Nivel Nodos 2021-01-31.csv

El archivo recibido es de tipo binario, se debe guardar en la carpeta una copia de extensión csv

```python
path_file = 'DATA/archivos_bajas_2021/'
archivos_bajas = os.listdir(path_file)

for j ,file in enumerate(archivos_bajas):
    mes = file[21:28]
    print(mes)
    # columna de producto homologado en algunos archivos vienen diferentes nombres
    try:
        use_cols = ['COD_NODO','SEGMENTO','Bajas Voluntarias']
        prod = 'Producto_Homologado'
        use_cols.append(prod)
        df = pd.read_csv(path_file+file, sep=';', decimal=',',  usecols=use_cols)
    except:
        use_cols = ['COD_NODO','SEGMENTO','Bajas Voluntarias']
        prod = 'PRODUCTO_HOMOLOGADO'
        use_cols.append(prod)
        df = pd.read_csv(path_file+file, sep=';', decimal=',',  usecols=use_cols)
    
    # seleccionando producto homologado internet y segmento residencial
    mask_arroba = (df[prod]=='@')
    mask_segmento = (df['SEGMENTO']=='RESIDENCIAL') | (df['SEGMENTO']=='RES')
    df = df[(mask_arroba & mask_segmento)].copy()
    
    # volviendo 0 los valores positivos en las bajas
    #df['Bajas Voluntarias'] = (df['Bajas Voluntarias']<0)*df['Bajas Voluntarias']
    
    # se realiza agrupación por nodos
    df_var = df.groupby('COD_NODO').agg({'Bajas Voluntarias':'sum'}).reset_index()
    df_var.columns = ['Nodo',mes]
    if j==0:
        df_var_2021 = df_var.copy()
    else:
        df_var_2021 = df_var_2021.merge(df_var.copy(), on ='Nodo', how='outer')


df_var_2021.set_index('Nodo', inplace=True)
df_var_2021 = (df_var_2021<0)*df_var_2021
df_var_2021.describe()
```

# Leyendo archivos de nodos segmentados

Los archivos de nodos segmentados desde mayo deben tener nombres con la siguiente estructura:

Nodos_Segmentados_Mayo_21.csv   (EL NOMBRE DEL MES DEBE ESTAR COMPLETO VER dic_meses_seg EN EL SIGUIENTE BLOQUE DE CODIGO

Si el archivo remitido por Jorge Saenz tiene extensión xlsx convertirlo a csv

```python
path_file = 'DATA/archivos_nodos_segmentados_2021/'
dic_meses_seg = {'ENERO':202101,'FEBRERO':202102,'MARZO':202103,'ABRIL':202104, 'BACK MARZO':202104,'MAYO':202105,
                'JUNIO':202106, 'JULIO':202107, 'AGOSTO':202108, 'SEPTIEMBRE':202109, 'OCTUBRE':202110, 
                 'NOVIEMBRE':202111, 'DICIEMBRE':202112}


df_segmentados_abril = pd.read_csv(f'{path_file}Segmentaciones_a_Abril_21.csv', sep=';', decimal=',')
df_segmentados_abril = df_segmentados_abril.explode('Nodo')
df_segmentados_abril = df_segmentados_abril[['Nodo','CRONOGRAMA VIGENTE']].copy()
df_segmentados_abril.columns = ['Nodo','mes']

# se agrega el nodo ERB que se había considerado en marzo pero no aparecía en el archivo hasta abril
df_segmentados_abril = df_segmentados_abril.append(pd.DataFrame.from_dict(data={'Nodo':['ERB'],'mes':['MARZO']}),
                                                   ignore_index=True)
df_segmentados_abril['mes'] = [dic_meses_seg[x] for x in df_segmentados_abril['mes']]
df_segmentados_abril = df_segmentados_abril[['Nodo','mes']]

list_df = [df_segmentados_abril]

path_file = path_file + 'desde_mayo/'
archivos_segmentados_desde_mayo = os.listdir(path_file)



for file in archivos_segmentados_desde_mayo:
    df_segmentados_mes = pd.read_csv(path_file + file, sep=';', decimal=',')
    df_segmentados_mes['Nodo'] = df_segmentados_mes['Nodo'].str.split(',')
    df_segmentados_mes = df_segmentados_mes.explode('Nodo')
    df_segmentados_mes = df_segmentados_mes[['Nodo']]
    mes = file[18:-7]
    print(mes)
    df_segmentados_mes['mes'] = dic_meses_seg[mes.upper()]
    list_df.append(df_segmentados_mes)
    
df_segmentados = pd.concat(list_df)
df_segmentados['mes'].value_counts()
```

# Leyendo archivo de Arpu y creando el dataframe de bajas

Copie todas las columnas del archivo segmentados_bajas.csv generado por el siguiente bloque de código y peguelo en la hoja data_bajas del archivo analisis_bajas_2021_comparado_ult_trimestre_2020.xlsx, el informe se genera en la hoja beneficios_bajas de este último archivo

```python
df_arpu = pd.read_csv('DATA/arpu.csv', sep=';', decimal=',')
df_arpu = df_arpu[['NODO','Arpu']].copy()
df_arpu.columns=['Nodo','Arpu']

df_excel = df_segmentados.merge(df_var_2020, on='Nodo', how='left')
df_excel = df_excel.merge(df_var_2021, on='Nodo', how='left')

list_meses_2021 = ['2021-01','2021-02','2021-03','2021-04','2021-05','2021-06','2021-07',
                  '2021-08','2021-09','2021-10','2021-11','2021-12']
df_excel.fillna(value=0, inplace=True)
for mes in list_meses_2021:
    if mes not in df_excel.columns:
        df_excel[mes] = 0
    
df_excel = df_excel.merge(df_arpu, on='Nodo', how='left')



df_excel.to_excel('segmentados_bajas.xlsx', index=False)
```

# Leyendo marcaciones

El archivo Llamadas_N1.csv debe ser actualizado para cada corrida con las llamadas desde  enero  de 2020 hasta el último mes finalizado.

```python
df_marcaciones = pd.read_csv('DATA/Llamadas_N1.csv', sep=';', decimal=',', usecols = ['NODO','ANIO','MES','CANT_LLAM_CUENTA'])
df_marcaciones.rename(columns={'NODO':'Nodo'}, inplace=True)
df_marcaciones['MES'] = [f'0{mes}' if mes<10 else f'{mes}' for mes in df_marcaciones['MES']]
df_marcaciones['MES'] = df_marcaciones['ANIO'].astype('str') + '_' + df_marcaciones['MES'] 
df_marcaciones = pd.pivot_table(df_marcaciones, values='CANT_LLAM_CUENTA', index='Nodo', columns='MES')
df_marcaciones.reset_index(inplace= True)
df_marcaciones.head()
```

```python
df_excel_marcaciones = df_segmentados.merge(df_marcaciones, on='Nodo', how='left')

df_excel_marcaciones.fillna(value=0, inplace=True)
list_meses_2021 = ['2021_01','2021_02','2021_03','2021_04','2021_05','2021_06','2021_07',
                  '2021_08','2021_09','2021_10','2021_11','2021_12']
for mes in list_meses_2021:
    if mes not in df_excel_marcaciones.columns:
        df_excel_marcaciones[mes] = 0
list_columns = ['Nodo','mes']
for a in [0,1]:
    for i in range(1,13):
        an_mes = f'202{a}_0{i}' if i <10 else f'202{a}_{i}'
        list_columns.append(an_mes)
df_excel_marcaciones = df_excel_marcaciones[list_columns].copy()
df_excel_marcaciones['VALOR'] = 3600
df_excel_marcaciones.to_excel('segmentados_marcaciones.xlsx', index=False)
```
